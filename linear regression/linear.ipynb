```python
from sklearn import model_selection     # data categoryy
from sklearn import datasets            # for dataset in scikit-learn
```

```python
#calling data split library
from sklearn .model_selection import train_test_split
labels=df['Sales']
features=df[['TV','Newspaper','Radio']]
```

```python
print(labels)
print(features)
```

```python
#plotting of graph with subplots
fig,ax= plt.subplots(3,figsize=(6,6)) # we can use in matplot lib using ax list ax[0]
#seaborn helps in building a statistical approach 
sns.pairplot(df,x_vars=['TV','Newspaper','Radio'],y_vars=['Sales'],kind='scatter')
plt.show
```

```python
# X_train,x_test,Y_train,y_test=train_test_split (features,labels,test_size=0.2,)
# X = training (80% training)         
# x = testing  (20% testing)
# Y = training (80% training)
# y = testing  (20% testing)
# Random state specifies the exact data that is taken by both the parties for training and testing
X_train,x_test,Y_train,y_test = train_test_split(features,labels,train_size=0.8)
# ml is an example of pipeline engineering 
```

```python
# where 1st and 3rd will be given to model/ algo
# calling linear model
model=LinearRegression()
```

```python
#passing training data to model
trained_model=model.fit(X_train,Y_train)
```

```python
#time to predict by passing testing data
predicted_output=trained_model.predict(x_test)
print('predicted_output',predicted_output)
time.sleep(2)
print ('Actual Output',y_test)
```
