```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/redashu/Datasets/master/advertising.csv'
```

```python
df = pd.read_csv(url)
```

```python
print(df.head())
```

```python
News=df['Newspaper'].tolist()
Rad=df['Radio'].tolist()
t_v=df['TV'].tolist()
```

```python
plt.xlabel('News')            
plt.ylabel('Rad')      
plt.title('Advertising Data')
plt.subplot(3,1,1)
print(plt.scatter(News,Rad,marker='*',color='red'))
plt.xlabel('t_v')            
plt.ylabel('Rad')      
plt.title('Advertising Data')
plt.subplot(3,1,2)
print(plt.scatter(Rad,t_v,marker='.',color='black'))
plt.xlabel('News')            
plt.ylabel('t_v')      
plt.title('Advertising Data')
plt.subplot(3,1,3)
print(plt.scatter(t_v,Rad,marker='^',color='blue'))
plt.show()
```
