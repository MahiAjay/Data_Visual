```python
# Decision tree classifier :

# -->The final target of decision tree classifier is to make a decision tree.
# -->The first node is the root node.
# -->The nodes which says no and do not have branches further are called leaf/termination node.
# -->The nodes which are further classified and have branches further are called decision nodes.

# -->There are 2 framework that are used to implement the decision tree class :
#                                                       cart  (used if you are using sklearn)
#                                                       ID-3  (used if you are using google)
#                     TERMS OF DECISION TREE CLASSIFIER
# --> Entropy : Used on the label which has to be the dataset's output on which is based
# --> I.G (Information Gain) : It is taken of the individual attribute 
#                              and then all the I.G is summed up.
# --> Gain : The I.G of every attribute is substracted from 1 and then compared.
#            The gratest gain amongst the attribute decides the root node.



# NOTE:
# To unimport or remove a library :
#     EG: import import pandas as pd
#     To unimport del pandas
```

```python
# loading modula
from sklearn.datasets import load_iris
type(load_iris)
```

```python
# loading data
iris = load_iris()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree  #for creating tree in visual form
```

```python
# Convert the data and target arrays into a pandas DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target variable as a new column
iris_df['target'] = iris.target

# Display the DataFrame
iris_df
```

```python
print(iris.target_names) #labels name
print(iris.feature_names) #features name
print(iris.data) # actual data of features
```

```python
features = iris.data
labels = iris.target
```

```python
#only setosa
setosa = features[:50]
plt.xlabel('Sepal_petal_Length')
plt.ylabel('Sepal_petal_Width')
plt.title('Setosa Data')
plt.scatter(setosa[:,0],setosa[:,1],c='red')
plt.scatter(setosa[:,2],setosa[:,3],c='green',marker='*')
plt.show()
```

```python
X_train,x_test,Y_train,y_test = train_test_split(features,labels,test_size=0.2)
```

```python
# calling model
model=DecisionTreeClassifier()
desc_trained=model.fit(X_train,Y_train)
```

```python
predicted_data = desc_trained.predict(x_test)
predicted_data
```

```python
#actual output
y_test
```

```python
#finding accuracy score
score = accuracy_score(y_test,predicted_data)
print('My accuracy of Decision Tree is: ',score)
```

```python
# plotting tree structure
plt.figure(figsize=(12,8)) # defining figure size
plot_tree(desc_trained ,feature_names=iris.feature_names,class_names=iris.target_names.tolist(),filled=True)
plt.title('Decision Tree')
plt.show()
```
